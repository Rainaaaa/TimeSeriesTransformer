```python
from dataset import calSharpeRatio
import pandas as pd
import joblib
import numpy as np

scaler = joblib.load("ret_scaler_single.obj")

predData = pd.read_csv("./output/Predictions_231120_153154.csv")
pred_char_Data = pd.read_csv("./output/Pred_test.csv")
risk_free_df = pd.read_csv("../v1/F-F_Research_Data_Factors.CSV")
```

```python
excessret = np.array(predData["ret_7"]).reshape(-1,1)
expred = np.array(predData["pred_ret_7"]).reshape(-1,1)

predData["ret_7_raw"] = scaler.inverse_transform(excessret).reshape(-1)
predData["pred_ret_7_raw"] = scaler.inverse_transform(expred).reshape(-1)

predData["pred_ret_7_raw"] = predData["pred_ret_7_raw"] / 100
risk_free_df["RF"] = risk_free_df["RF"] / 100

# process charateristics data
pred_char_Data = pred_char_Data.drop(["month_1", "month_2", "month_3", "month_4", "month_5", "month_6"], axis=1)
pred_char_Data.rename(columns={"month_0": "month"}, inplace=True)

# concatenate the prediction and charateraistics data
predData = pd.concat([predData, pred_char_Data], axis=1)
```

```python
months = list(predData.month.unique())
months.sort()
```

```python
s, months = calSharpeRatio(months=months, rfdf=risk_free_df, df=predData, charac= "pred_ret_7_raw", col="pred_ret_7_raw")
```

```python
# sharp ratio
print("sharp ratio:", s)
print("monthly means:", months)
```
